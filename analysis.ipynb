```python
import re
```

```python
def get_int(value):
    temp= re.findall(r'\d+', value)
    return temp[0]

new_df["AccommodationCost"]= new_df["Accommodation cost"].apply(lambda x:get_int(x))
new_df["TransportationCost"]= new_df["Transportation cost"].apply(lambda x:get_int(x))
```

```python
new_df.head()
```

```python
new_df.drop(["Transportation cost", "Accommodation cost"], axis= 1, inplace= True)
```

```python
new_df= new_df.astype({"AccommodationCost": 'int', "TransportationCost": 'int', "Duration (days)": 'int'})
```

```python
new_df.info()
```

```python
new_df["TotalAccommodationCost"]= new_df["AccommodationCost"]*new_df["Duration (days)"]
new_df["TotalTransportationCost"]= new_df["TransportationCost"]*new_df["Duration (days)"]
new_df["TotalTravelCost"]= new_df["TotalAccommodationCost"]+new_df["TotalTransportationCost"]
```
